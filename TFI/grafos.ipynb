```python
import networkx as nx

grafo = nx.DiGraph()
grafo.add_edges_from([(1,3),(1,2),(2,3),(2,5),(3,4),(3,8),(4,1),(4,5),(5,2),(5,3),(6,5),(6,7),(7,2),(7,3),(8,1),(8,6)])

print("Vertices del grafo son: {}".format(list(grafo.nodes)))
print("Aristas del grafo son: {}".format(list(grafo.edges)))
print("")

print("Elementos vecinos de 1 en el grafo son {}".format(list(grafo.neighbors(1))))
print("Elementos vecinos de 2 en el grafo son {}".format(list(grafo.neighbors(2))))
print("Elementos vecinos de 3 en el grafo son {}".format(list(grafo.neighbors(3))))
print("Elementos vecinos de 4 en el grafo son {}".format(list(grafo.neighbors(4))))
print("Elementos vecinos de 5 en el grafo son {}".format(list(grafo.neighbors(5))))
print("Elementos vecinos de 6 en el grafo son {}".format(list(grafo.neighbors(6))))
print("Elementos vecinos de 7 en el grafo son {}".format(list(grafo.neighbors(7))))
print("Elementos vecinos de 8 en el grafo son {}".format(list(grafo.neighbors(8))))
print("")

nx.draw(grafo)
```
