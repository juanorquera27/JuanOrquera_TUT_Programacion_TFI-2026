```python
import networkx as nx
import matplotlib.pyplot as plt
from typing import Dict, List

class RedDireccionada:
    """
    Representa una topología de red mediante un grafo dirigido con soporte visual.
    """
    def __init__(self, total_nodos: int, conexiones_por_nodo: int) -> None:
        self.total_nodos = total_nodos
        self.conexiones_por_nodo = conexiones_por_nodo
        self.grafo: Dict[int, List[int]] = self._construir_topologia()

    def _construir_topologia(self) -> Dict[int, List[int]]:
        """Construye la lista de adyacencia interna."""
        red: Dict[int, List[int]] = {}
        for i in range(self.total_nodos):
            enlaces = [(i + salto) % self.total_nodos for salto in range(1, self.conexiones_por_nodo + 1)]
            red[i] = enlaces
        return red

    def renderizar_estado(self) -> None:
        """Imprime la configuración en la consola."""
        print("--- TOPOLOGÍA DE RED DIRECCIONADA ---")
        for nodo, enlaces in self.grafo.items():
            destinos_str = " y ".join([f"Nodo {d}" for d in enlaces])
            print(f"[Nodo {nodo}] transmite directamente hacia -> {destinos_str}")

    def graficar_red(self) -> None:
        """
        Genera y despliega una representación visual interactiva del grafo.
        """
        print("\nIniciando motor gráfico (Matplotlib/NetworkX)...")

        # se instancia el grafo dirigido
        G = nx.DiGraph()

        # Alimentamos el objeto visual con nuestra lista de adyacencia
        for origen, destinos in self.grafo.items():
            for destino in destinos:
                G.add_edge(origen, destino)

        # Configuramos la física y la estética del grafo
        plt.figure(figsize=(8, 8))

        # circular_layout posiciona los nodos en un anillo perfecto
        posiciones = nx.circular_layout(G)

        # Renderizamos los nodos y las flechas
        nx.draw(
            G,
            pos=posiciones,
            with_labels=True,
            node_color='darkcyan',
            font_color='white',
            font_weight='bold',
            node_size=2000,
            arrowsize=20,
            edge_color='gray'
        )

        plt.title(f"Red Direccionada: {self.total_nodos} Nodos", fontsize=14, fontweight='bold')

        # Desplegamos la ventana de UI
        plt.show()

# entrada principal
if __name__ == "__main__":
    sistema_nodos = RedDireccionada(total_nodos=8, conexiones_por_nodo=2)

    # Renderizamos primero en texto (Logs)
    sistema_nodos.renderizar_estado()

    # Lanzamos la interfaz gráfica (UI)
    sistema_nodos.graficar_red()
```
